### Get Input Data

```python
## Read Data.
preprocessed_data_path = 'Data\\Preprocessed_Data'
encoded_labels_path = 'Data\\Preprocessed_Data\\train_label\\train_label-encoded.csv'
raw_data_train_image_path = 'Data\\Raw_Data\\train_image'

df_labels = pd.read_csv(encoded_labels_path)

# Rename columns..
# df_label.columns = ['Image_Name',1,2,3,4]

# Cut it to First 15 Rows for now..
# Never Mind...
# df_labels = df_labels[:15000]
df_labels
```

```python
# Remove Dir..(testing..)
shutil.rmtree(preprocessed_data_path+'\\train_images\\')
shutil.rmtree(preprocessed_data_path+'\\validation_images\\')
```

```python
def create_classification_folders():
    
    # Create for all classes..
    train_dirs = [
        preprocessed_data_path+'\\train_images\\0',
        preprocessed_data_path+'\\train_images\\1',
        preprocessed_data_path+'\\train_images\\2',
        preprocessed_data_path+'\\train_images\\3',
        preprocessed_data_path+'\\train_images\\4',
        preprocessed_data_path+'\\train_images\\5',
        preprocessed_data_path+'\\train_images\\6',
        preprocessed_data_path+'\\train_images\\7',
        ]
    validation_dirs = [
        preprocessed_data_path+'\\validation_images\\0',
        preprocessed_data_path+'\\validation_images\\1',
        preprocessed_data_path+'\\validation_images\\2',
        preprocessed_data_path+'\\validation_images\\3',
        preprocessed_data_path+'\\validation_images\\4',
        preprocessed_data_path+'\\validation_images\\5',
        preprocessed_data_path+'\\validation_images\\6',
        preprocessed_data_path+'\\validation_images\\7',
        ]
    
    os.makedirs(preprocessed_data_path+'\\train_images',exist_ok=True)
    os.makedirs(preprocessed_data_path+'\\validation_images',exist_ok=True)

    for dir in train_dirs:
        os.makedirs(dir,mode=0o777,exist_ok=True)
    for dir in validation_dirs:
        os.makedirs(dir,mode=0o777,exist_ok=True)

    return train_dirs, validation_dirs

class_folder_train_dirs, class_folder_validation_dirs = create_classification_folders()
```

```python
import random

random.seed = 7

def threshold_randomizer(threshold):
    """ Less than the threshold returns True.."""
    if random.random() < threshold:
        return True
    else:
        return False
    #print(class_counter)
    #print(f'{class_counter[0]/np.sum(class_counter)} , {class_counter[1]/np.sum(class_counter)}')
```

```python
# Copy images to respective folders..
# TRAIN_VALIDATION_SPLIT = 0.975
TRAIN_VALIDATION_SPLIT = 0.9875
# 80% Train.. 
# 20% Test..

for idx, row in df_labels.iterrows():
    # Get the image..
    raw_img_path = 'Data\\Raw_Data\\train_image\\labeled_data\\'+row['Image_Name']
    
    print(f'Processing: {idx} / {df_labels.shape[0]}')

    # copy_path = 
    to_insert_folders = list(row[row==1].index)

    # Threshold Randomizer is put outside, so that the same image will not make it to both Train and Validation sets..
    is_in_train_set = threshold_randomizer(TRAIN_VALIDATION_SPLIT)

    #print(to_insert_folders)
    if(len(to_insert_folders)==0):
        # Copy it directly to 0 folder..
        print(f'Copying File {row["Image_Name"]} to folder 0')
        if is_in_train_set:
            shutil.copy(raw_img_path, class_folder_train_dirs[0])
        else:
            shutil.copy(raw_img_path, class_folder_validation_dirs[0])
    else:
        
        for folder in to_insert_folders:
            # print(folder)
            #print(class_folder_dirs[int(folder)-1])
            #print(raw_img_path,class_folder_dirs[int(folder)-1])
            print(f'Copying File {row["Image_Name"]} to folder {folder}')
            if is_in_train_set:
                shutil.copy(raw_img_path, class_folder_train_dirs[int(folder)])
            else:
                shutil.copy(raw_img_path, class_folder_validation_dirs[int(folder)])
```
